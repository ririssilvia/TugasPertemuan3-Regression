### Visualisasi Data

```python
# import library untuk visualisasi
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# operasi visualisasi
def pp(x,y,z):
    sns.pairplot(data, x_vars=[x,y,z], y_vars='price',size=4, aspect=1, kind='scatter')
    plt.show()

pp('enginesize', 'boreratio', 'stroke')
pp('compressionratio', 'horsepower', 'peakrpm')
pp('wheelbase', 'citympg', 'highwaympg')
```

output di atas kurang bisa menunjukkan korelasi antar data dalam x dengan data dalam y. Salah satu solusinya adalah menggunakan heatmap

```python

plt.figure(figsize = (30, 25))
sns.heatmap(data.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

dari bentuk visualisasi di atas terlihat bahwa curbweight, enginesize, horsepower,carwidth and highend memiliki korelasi yang paling kuat terhadap  price.

### Regresi Linier

```python
# Buat variabel bebas X dan Y, sebagai contoh ambil dari hasil analisis korelasi dari kegaitan sebelumnya
X = data['horsepower']
y = data['price']
```

```python
# Buat pemisahan data uji dan data latih dengan proporsi 7:3
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, 
                                                    test_size = 0.3, random_state = 100)
```

```python
# hasil training dtaset
X_train
y_train
```

```python
# training model
import statsmodels.api as sm

X_train_sm = sm.add_constant(X_train)
```

```python
# fitting garis regresi
lr = sm.OLS(y_train, X_train_sm).fit()
lr.params
```

```python
# analisis statistika dari garis regresi
lr.summary()
```

```python
# visualisasi garis regresi
plt.scatter(X_train, y_train)
plt.plot(X_train, -3192.650565 + 158.445735*X_train, 'r')
plt.show()
```

### Residual Analysis
dipakai untuk mengetahui tingkat error dari variabel yang dipengaruhi (y)

Error = Actual y value - y predicted value

```python
# prediksi y_value dari data x yang telah dilatih
y_train_pred = lr.predict(X_train_sm)

res = (y_train - y_train_pred)
```

```python
# cek histogram apakah berdistribusi normal atau tidak
fig = plt.figure()
sns.distplot(res, bins = 15)
plt.title('Error Terms', fontsize = 15)
plt.xlabel('y_train - y_train_pred', fontsize = 15)
plt.show()
```

```python
plt.scatter(X_train,res)
plt.show()
```

```python
# prediksi pada data uji dan evaluasi model
X_test_sm = sm.add_constant(X_test)

# prediksi y value yang berkorelasi dengan X_test_sm
y_test_pred = lr.predict(X_test_sm)

# cetak 5 data terprediksi teratas
y_test_pred.head()
```

```python
# hitung nilai r^2
from sklearn.metrics import r2_score

r_squared = r2_score(y_test, y_test_pred)
r_squared
```

perlu diperhatikan bahwa r^2 dari data trainign adalah 0.650 sedangkan dari data testing adalah 0.655. Hal ini berarti model yang dibentuk cukup stabil (tidak berselisih jauh antara training dengan testing)

```python
# visualisasi data
plt.scatter(X_test, y_test)
plt.plot(X_test, y_test_pred, 'r')
plt.show()
```
